## Combining Data

```python
city_data1 = pd.DataFrame({
    'revenue':[7000000,2000000],
    'employee':[800,900]},
    index= ['Karawang',"Semarang"]
    )
city_data1
```

```python
city_data
```

```python
all_city = pd.concat([city_data,city_data1], sort=False)
all_city
```

```python
city_countries = pd.DataFrame({
    "country":['Holland','Japan','Holland','Canada','Spain'],
    "is_capital":[1,1,0,0,0]
}, index=['Amsterdam','Tokyo','Rotterdam','Toronto','Barcelona'])

city_countries
```

```python
city_GDP = pd.DataFrame({
    "revenue":[2500,2000,3000,2500,2400],
    "employee_count":[1,2,3,4,5]
}, index=['Amsterdam','Tokyo','Rotterdam','Toronto','Barcelona'])

city_GDP
```

```python
cities = pd.concat([city_countries,city_GDP], sort=False, axis=1)
cities
```

```python
countries = pd.DataFrame({
    'population_millions':[17,127,37],
    'continent':['Europe','Asia','North America']
}, index=['Holland','Japan','Canada'])
countries
```

```python
new = pd.merge(cities,countries, left_on='country',right_index=True)
new
```

```python
pd.merge(cities,countries, left_on='country',right_index=True,how='left')
```
